```python
from pyecharts.globals import CurrentConfig, NotebookType
CurrentConfig.NOTEBOOK_TYPE = NotebookType.JUPYTER_LAB

from pyecharts.charts import Line
from pyecharts import options as opts
from pyecharts.globals import ThemeType

import pandas as pd
```

## Thrift Performance

```python
thrift_df = pd.read_csv('thrift.csv').set_index('size')
thrift_df = thrift_df.transpose()
```

```python
line = Line(init_opts=opts.InitOpts(theme=ThemeType.DARK, width='100%'))
line.set_global_opts(
    yaxis_opts=opts.AxisOpts(
        type_="log",
        name="time",
        splitline_opts=opts.SplitLineOpts(is_show=True),
        is_scale=True,
        min_=1E-4,
        max_=10,
    ),
    xaxis_opts=opts.AxisOpts(
        name="size",
    ),
    tooltip_opts=opts.TooltipOpts(
        trigger='axis',
    )
)
line.add_xaxis(thrift_df.index.tolist())
for col in thrift_df.columns:
    line.add_yaxis(
        col, thrift_df[col].tolist(),
        label_opts=opts.LabelOpts(
            is_show=False
        )
    )
line.load_javascript()
```

```python
line.render_notebook()
```
